```python
import pandas as pd
from sqlalchemy import create_engine
from config import username, password 
```

```python
csv_county = "Data/county_data_2017.csv"
county_df = pd.read_csv(csv_county)
county_df.head()
```

```python
county_df = county_df[['State', 'County', 'TotalPop', 'Income', 'IncomeErr', 'IncomePerCap', 'IncomePerCapErr', 'Poverty', 'Unemployment']].copy()
county_df = county_df.rename(columns = {'State':'state', 'County':'county', 'TotalPop':'total_pop', 
                            'Income':'income', 'IncomeErr':'income_err', 'IncomePerCap':'income_per_cap', 
                            'IncomePerCapErr':'income_per_cap_err', 'Poverty':'poverty', 'Unemployment':'unemployment'}).copy()
county_df.head()
```

```python
csv_vaccine = "Data/Vaccine_Hesitancy.csv"
vaccine_df = pd.read_csv(csv_vaccine)
vaccine_df.head()
```

```python
vaccine_df['County Name'] = vaccine_df["County Name"].str.split(",", n = 1, expand = True)

vaccine_df = vaccine_df[['County Name', 'State', 'Estimated hesitant',
       'Estimated strongly hesitant', 'Social Vulnerability Index (SVI)',
       'CVAC level of concern for vaccination rollout',
       'Percent adults fully vaccinated against COVID-19']].copy()
vaccine_df = vaccine_df.rename(columns = {'County Name':'county_name', 'State':'state', 'Estimated hesitant':'est_hesitant',
       'Estimated strongly hesitant': 'est_strongly_hesitant', 'Social Vulnerability Index (SVI)': 'svi',
       'CVAC level of concern for vaccination rollout': 'cvac_level_of_concern_rollout',
       'Percent adults fully vaccinated against COVID-19':'percent_adults_fully_vaccinated'}).copy()

vaccine_df['state'] = vaccine_df['state'].str.title().copy()

vaccine_df.head()
```

```python
#  Connect to local database

rds_connection_string = f"{username}:{password}@localhost:5432/Project2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names() 
```

```python
# Use pandas to load csv converted DataFrame into database for census data
county_df.to_sql(name='census_data', con=engine, if_exists='append', index=False) 
```

```python
# Use pandas to load csv converted DataFrame into database for vaccine hesitancy data
vaccine_df.to_sql(name='vaccine_data', con=engine, if_exists='append', index=False) 
```

```python
# Display final table
pd.read_sql_query('SELECT * FROM census_vaccine_data', con=engine)
```
